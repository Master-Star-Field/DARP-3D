```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import random
import math
import os
from collections import deque
import time
from tqdm import tqdm
import matplotlib.patches as patches
import sys
from scipy.spatial import cKDTree
import time
import numba as nb
from numba import jit, prange

# Класс для среды обучения подводного дрона
class AUVEnvironment:
    def __init__(self):
        # Константы модели AUV (увеличены для более быстрого движения)
        self.auv_speed = 3.5  # Увеличено с 1.0 до 3.5 узлов
        self.delta_t = 0.2    # Увеличено с 0.1 до 0.2 часов
        self.num_sonars = 12  
        self.sonar_angle = 10 
        self.sonar_range = 5.0  # Увеличено с 3.0 до 5.0 морских миль
        self.num_actions = 16  
        
        # Параметры среды
        self.area_width = 100  
        self.area_height = 70  
        self.num_obstacles = 11  
        self.obstacle_radius = 3.0  
        self.start_pos = np.array([90.0, 5.0])  
        self.goal_pos = np.array([20.0, 60.0])  
        
        # Генерация препятствий
        self.obstacles = self._generate_obstacles()
        
        # Создание KD-дерева для быстрого поиска ближайших препятствий
        self.obstacle_tree = cKDTree(self.obstacles)
        
        # Предварительный расчет направлений сонаров
        self.sonar_directions = np.zeros((self.num_sonars, 2))
        
        # Генерация поля течений
        self.current_field = self._generate_current_field()
        self._precompute_current_grid()
        
        # Текущее состояние
        self.position = np.copy(self.start_pos)
        self.heading = 0.0
        self.prev_action = None
        self.steps = 0
        self.max_steps = 2000  # Увеличено для дольших эпизодов
        
        # Траектория для визуализации
        self.trajectory = [np.copy(self.position)]
        
        # Кэш для оптимизации повторяющихся расчетов
        self._cached_position = None
        self._cached_sonar_readings = None
        self._cached_current = None
        
        # Предварительный расчет углов сонаров для избегания повторных вычислений
        self._precompute_sonar_angles()
    
    def _precompute_sonar_angles(self):
        """Предварительный расчет углов сонаров для всех возможных курсов"""
        # Преобразование угла сонара в радианы
        sonar_angle_rad = np.radians(self.sonar_angle)
        
        # Расчет относительных углов для всех сонаров
        self.sonar_relative_angles = np.array([
            (i - (self.num_sonars - 1) / 2) * sonar_angle_rad 
            for i in range(self.num_sonars)
        ])
        
        # Предварительно рассчитываем направление для сонаров при начальном курсе
        self._update_sonar_directions()
    
    def _update_sonar_directions(self):
        """Обновление направлений сонаров при изменении курса"""
        for i, rel_angle in enumerate(self.sonar_relative_angles):
            angle = self.heading + rel_angle
            self.sonar_directions[i] = np.array([np.cos(angle), np.sin(angle)])
    
    def _precompute_current_grid(self):
        """Предварительный расчет индексов сетки течений для оптимизации интерполяции"""
        cf = self.current_field
        # Расчет шага сетки
        self.dx = (cf['x_grid'][-1] - cf['x_grid'][0]) / (len(cf['x_grid']) - 1)
        self.dy = (cf['y_grid'][-1] - cf['y_grid'][0]) / (len(cf['y_grid']) - 1)
        
        # Инверсия шага для быстрого расчета индексов
        self.inv_dx = 1.0 / self.dx
        self.inv_dy = 1.0 / self.dy
        
        # Начальные координаты сетки
        self.x0 = cf['x_grid'][0]
        self.y0 = cf['y_grid'][0]
        
        # Размеры сетки
        self.nx = len(cf['x_grid'])
        self.ny = len(cf['y_grid'])
    
    # Оптимизированный метод получения течения в точке
    def _get_current_at_position(self, pos):
        """Оптимизированное получение значения течения через билинейную интерполяцию"""
        # Проверка кэша
        if self._cached_position is not None and np.array_equal(pos, self._cached_position):
            return self._cached_current
        
        x, y = pos
        cf = self.current_field
        
        # Быстрый расчет индексов вместо поиска
        ix = int((x - self.x0) * self.inv_dx)
        iy = int((y - self.y0) * self.inv_dy)
        
        # Проверка границ
        ix = max(0, min(ix, self.nx - 2))
        iy = max(0, min(iy, self.ny - 2))
        
        # Расчет весов для интерполяции
        wx = (x - (self.x0 + ix * self.dx)) * self.inv_dx
        wy = (y - (self.y0 + iy * self.dy)) * self.inv_dy
        
        # Билинейная интерполяция
        u00 = cf['u'][iy, ix]
        u01 = cf['u'][iy, ix+1]
        u10 = cf['u'][iy+1, ix]
        u11 = cf['u'][iy+1, ix+1]
        
        v00 = cf['v'][iy, ix]
        v01 = cf['v'][iy, ix+1]
        v10 = cf['v'][iy+1, ix]
        v11 = cf['v'][iy+1, ix+1]
        
        u = (1-wx)*(1-wy)*u00 + wx*(1-wy)*u01 + (1-wx)*wy*u10 + wx*wy*u11
        v = (1-wx)*(1-wy)*v00 + wx*(1-wy)*v01 + (1-wx)*wy*v10 + wx*wy*v11
        
        # Сохраняем в кэш
        self._cached_position = np.copy(pos)
        self._cached_current = np.array([u, v])
        
        return self._cached_current
    
    # Оптимизированный метод для получения показаний сонаров
    def _get_sonar_readings(self):
        """Векторизованное получение показаний сонаров"""
        # Обновляем направления сонаров при изменении курса
        self._update_sonar_directions()
        
        # Инициализация показаний максимальной дальностью
        readings = np.ones(self.num_sonars) * self.sonar_range
        
        # Находим ближайшие препятствия в радиусе действия сонаров
        # для оптимизации проверяем только потенциально видимые препятствия
        max_check_distance = self.sonar_range + self.obstacle_radius
        nearby_indices = self.obstacle_tree.query_ball_point(
            self.position, max_check_distance)
        
        # Если нет близких препятствий, возвращаем максимальную дальность
        if not nearby_indices:
            return readings
        
        # Фильтруем только ближайшие препятствия
        nearby_obstacles = self.obstacles[nearby_indices]
        
        # Векторизованный расчет показаний сонаров
        return self._vectorized_sonar_calculation(nearby_obstacles, readings)
    
    @staticmethod
    @jit(nopython=True, parallel=True)
    def _vectorized_sonar_calculation(position, directions, obstacles, obstacle_radius, sonar_range, readings):
        """JIT-компилированная функция для расчета показаний сонаров"""
        num_sonars = len(directions)
        
        for i in prange(num_sonars):
            direction = directions[i]
            
            for j in range(len(obstacles)):
                obstacle_pos = obstacles[j]
                
                # Вектор от позиции AUV до центра препятствия
                to_obstacle = obstacle_pos - position
                
                # Проекция этого вектора на направление сонара
                projection = to_obstacle[0] * direction[0] + to_obstacle[1] * direction[1]
                
                # Если проекция отрицательная, препятствие за AUV
                if projection < 0:
                    continue
                
                # Квадрат расстояния от AUV до препятствия
                dist_sq = to_obstacle[0] * to_obstacle[0] + to_obstacle[1] * to_obstacle[1]
                
                # Квадрат проекции
                proj_sq = projection * projection
                
                # Расстояние от прямой сонара до центра препятствия
                closest_dist = np.sqrt(max(0.0, dist_sq - proj_sq))
                
                # Если препятствие достаточно близко к линии сонара и в пределах его дальности
                if closest_dist <= obstacle_radius and projection <= sonar_range:
                    # Вычисляем точку пересечения сонара с препятствием
                    intersect_dist = projection - np.sqrt(max(0.0, obstacle_radius*obstacle_radius - closest_dist*closest_dist))
                    
                    # Обновляем показание сонара, если это расстояние меньше текущего
                    if 0 <= intersect_dist < readings[i]:
                        readings[i] = intersect_dist
        
        return readings
    
    def _vectorized_sonar_calculation(self, nearby_obstacles, readings):
        """Обертка для вызова JIT-компилированной функции"""
        return self._jit_sonar_calculation(
            self.position, self.sonar_directions, nearby_obstacles, 
            self.obstacle_radius, self.sonar_range, readings
        )
    
    # Пользовательская реализация JIT-функции
    @staticmethod
    @jit(nopython=True)
    def _jit_sonar_calculation(position, directions, obstacles, obstacle_radius, sonar_range, readings):
        """JIT-компилированная функция для расчета показаний сонаров"""
        for i in range(len(directions)):
            direction = directions[i]
            
            for obstacle_pos in obstacles:
                # Вектор от позиции AUV до центра препятствия
                to_obstacle = obstacle_pos - position
                
                # Проекция этого вектора на направление сонара
                projection = np.dot(to_obstacle, direction)
                
                # Если проекция отрицательная, препятствие за AUV
                if projection < 0:
                    continue
                
                # Расстояние от прямой сонара до центра препятствия
                closest_dist = np.linalg.norm(to_obstacle - projection * direction)
                
                # Если препятствие достаточно близко к линии сонара и в пределах его дальности
                if closest_dist <= obstacle_radius and projection <= sonar_range:
                    # Вычисляем точку пересечения сонара с препятствием
                    intersect_dist = projection - np.sqrt(max(0.0, obstacle_radius**2 - closest_dist**2))
                    
                    # Обновляем показание сонара, если это расстояние меньше текущего
                    if 0 <= intersect_dist < readings[i]:
                        readings[i] = intersect_dist
        
        return readings
    
    # Оптимизированная проверка коллизий
    def _check_collision(self, position):
        """Быстрая проверка коллизий с использованием KD-дерева"""
        # Поиск ближайшего препятствия
        distance, _ = self.obstacle_tree.query(position, k=1)
        return distance <= self.obstacle_radius
    
    # Оптимизированный метод step
    def step(self, action):
        """Оптимизированное выполнение шага в среде"""
        t_start = time.time()  # Для отладки
        
        # Преобразование дискретного действия в угол
        action_angle = 2 * np.pi * action / self.num_actions
        
        # Вычисление вектора скорости AUV
        auv_velocity = self.auv_speed * np.array([
            np.cos(action_angle),
            np.sin(action_angle)
        ])
        
        # Получение вектора течения
        current_velocity = self._get_current_at_position(self.position)
        
        # Общий вектор перемещения
        total_velocity = auv_velocity + current_velocity
        
        # Обновление позиции
        new_position = self.position + total_velocity * self.delta_t
        
        # Быстрая проверка коллизий
        collision = self._check_collision(new_position)
        
        # Проверка выхода за пределы области
        out_of_bounds = (
            new_position[0] < 0 or new_position[0] > self.area_width or
            new_position[1] < 0 or new_position[1] > self.area_height
        )
        
        # Если нет коллизии и не выходим за границы, обновляем позицию
        if not collision and not out_of_bounds:
            self.position = new_position
            self.trajectory.append(np.copy(self.position))
            # Инвалидируем кэши
            self._cached_position = None
            self._cached_sonar_readings = None
        
        # Обновление курса
        if np.linalg.norm(total_velocity) > 0:
            self.heading = np.arctan2(total_velocity[1], total_velocity[0])
        
        # Вычисление награды
        reward = self._calculate_reward(action, collision, out_of_bounds)
        
        # Проверка достижения цели
        distance_to_goal = np.linalg.norm(self.position - self.goal_pos)
        goal_reached = distance_to_goal < 2.0
        
        # Инкремент счетчика шагов
        self.steps += 1
        
        # Проверка завершения эпизода
        done = goal_reached or collision or out_of_bounds or self.steps >= self.max_steps
        
        # Запоминаем предыдущее действие для награды за плавность
        self.prev_action = action
        
        # Создаем словарь с информацией о шаге
        info = {
            'goal_reached': goal_reached,
            'collision': collision,
            'out_of_bounds': out_of_bounds,
            'steps': self.steps,
            'position': self.position,
            'heading': self.heading,
            'distance': distance_to_goal,
            'step_time': time.time() - t_start  # Для отладки
        }
        
        return self.get_state(), reward, done, info
    
    
    def _generate_obstacles(self):
        """Генерация случайных препятствий с учетом начальной и целевой точек"""
        obstacles = []
        min_dist_to_points = 10.0  # минимальное расстояние от препятствий до начальной и целевой точек
        
        while len(obstacles) < self.num_obstacles:
            x = np.random.uniform(0, self.area_width)
            y = np.random.uniform(0, self.area_height)
            pos = np.array([x, y])
            
            # Проверка расстояния до начальной и целевой точек
            if (np.linalg.norm(pos - self.start_pos) < min_dist_to_points + self.obstacle_radius or
                np.linalg.norm(pos - self.goal_pos) < min_dist_to_points + self.obstacle_radius):
                continue
            
            # Проверка перекрытия с другими препятствиями
            overlap = False
            for obs in obstacles:
                if np.linalg.norm(pos - obs) < 2 * self.obstacle_radius:
                    overlap = True
                    break
            
            if not overlap:
                obstacles.append(pos)
        
        return np.array(obstacles)
    
    def _generate_current_field(self):
        """Генерация поля океанских течений"""
        # Создаем сетку 10x10 для интерполяции течений
        grid_size = 10
        
        # Генерируем значения u и v компонент течения для каждой точки сетки
        u_currents = np.random.uniform(-0.5, 0.5, (grid_size, grid_size))
        v_currents = np.random.uniform(-0.5, 0.5, (grid_size, grid_size))
        
        # Создаем координаты сетки
        x_grid = np.linspace(0, self.area_width, grid_size)
        y_grid = np.linspace(0, self.area_height, grid_size)
        
        return {
            'u': u_currents,
            'v': v_currents,
            'x_grid': x_grid,
            'y_grid': y_grid,
            'grid_size': grid_size
        }
    
    def _get_current_at_position(self, pos):
        """Получение значения течения в текущей позиции через билинейную интерполяцию"""
        x, y = pos
        cf = self.current_field
        
        # Находим индексы ближайших точек сетки
        i = np.searchsorted(cf['x_grid'], x) - 1
        j = np.searchsorted(cf['y_grid'], y) - 1
        
        # Обеспечиваем, что индексы находятся в пределах сетки
        i = max(0, min(i, cf['grid_size'] - 2))
        j = max(0, min(j, cf['grid_size'] - 2))
        
        # Вычисляем веса для интерполяции
        x1, x2 = cf['x_grid'][i], cf['x_grid'][i+1]
        y1, y2 = cf['y_grid'][j], cf['y_grid'][j+1]
        
        wx = (x - x1) / (x2 - x1) if x2 != x1 else 0
        wy = (y - y1) / (y2 - y1) if y2 != y1 else 0
        
        # Билинейная интерполяция компонентов течения
        u = (1-wx)*(1-wy)*cf['u'][j,i] + wx*(1-wy)*cf['u'][j,i+1] + (1-wx)*wy*cf['u'][j+1,i] + wx*wy*cf['u'][j+1,i+1]
        v = (1-wx)*(1-wy)*cf['v'][j,i] + wx*(1-wy)*cf['v'][j,i+1] + (1-wx)*wy*cf['v'][j+1,i] + wx*wy*cf['v'][j+1,i+1]
        
        return np.array([u, v])
    
    def _get_sonar_readings(self):
        """Получение показаний сонара - расстояний до препятствий"""
        readings = np.ones(self.num_sonars) * self.sonar_range
        
        # Преобразование угла сонара в радианы
        sonar_angle_rad = math.radians(self.sonar_angle)
        
        for i in range(self.num_sonars):
            # Вычисляем угол для текущего сонара (относительно курса AUV)
            angle = self.heading + (i - (self.num_sonars - 1) / 2) * sonar_angle_rad
            
            # Вектор направления сонара
            direction = np.array([math.cos(angle), math.sin(angle)])
            
            # Проверяем пересечение с каждым препятствием
            for obstacle_pos in self.obstacles:
                # Вектор от позиции AUV до центра препятствия
                to_obstacle = obstacle_pos - self.position
                
                # Проекция этого вектора на направление сонара
                projection = np.dot(to_obstacle, direction)
                
                # Если проекция отрицательная, препятствие за AUV
                if projection < 0:
                    continue
                
                # Расстояние от прямой сонара до центра препятствия
                closest_dist = np.linalg.norm(to_obstacle - projection * direction)
                
                # Если препятствие достаточно близко к линии сонара и в пределах его дальности
                if closest_dist <= self.obstacle_radius and projection <= self.sonar_range:
                    # Вычисляем точку пересечения сонара с препятствием
                    intersect_dist = projection - math.sqrt(max(0, self.obstacle_radius**2 - closest_dist**2))
                    
                    # Обновляем показание сонара, если это расстояние меньше текущего
                    if 0 <= intersect_dist < readings[i]:
                        readings[i] = intersect_dist
        
        return readings
    
    def get_state(self):
        """Получение вектора состояния для RL-агента"""
        # Относительный вектор до цели
        rel_goal = self.goal_pos - self.position
        
        # Получение данных о течениях
        current = self._get_current_at_position(self.position)
        
        # Получение показаний сонаров
        sonar_readings = self._get_sonar_readings()
        
        # Формирование вектора состояния
        state = np.concatenate([
            rel_goal,                # 2 элемента: относительный вектор до цели
            [self.heading],          # 1 элемент: текущий курс
            current,                 # 2 элемента: вектор течения
            sonar_readings           # 12 элементов: показания сонаров
        ])
        
        return state
    
    def reset(self):
        """Сброс среды в начальное состояние"""
        self.position = np.copy(self.start_pos)
        self.heading = 0.0
        self.prev_action = None
        self.steps = 0
        
        # Сброс траектории
        self.trajectory = [np.copy(self.position)]
        
        return self.get_state()
    
    def _calculate_reward(self, action, collision, out_of_bounds):
        """Расчет функции награды"""
        # Если столкновение или выход за границы - возвращаем большой штраф
        if collision or out_of_bounds:
            return -20
        
        # Компонент награды за расстояние до цели
        prev_distance = np.linalg.norm(self.position - self.goal_pos if len(self.trajectory) <= 1 
                                    else self.trajectory[-2] - self.goal_pos)
        distance_to_goal = np.linalg.norm(self.position - self.goal_pos)
        r_distance = prev_distance - distance_to_goal
        
        # Компонент награды за близость к препятствиям
        sonar_readings = self._get_sonar_readings()
        r_obstacles = 0
        for i, d in enumerate(sonar_readings):
            # Больше штрафуем за препятствия по курсу, чем по бокам
            weight = math.exp(-abs(i - 5.5))
            r_obstacles += (1 - d / self.sonar_range) * weight
        
        # Компонент награды за использование течений
        current = self._get_current_at_position(self.position)
        current_speed = np.linalg.norm(current)
        if current_speed > 0:
            action_angle = 2 * math.pi * action / self.num_actions
            action_vector = np.array([math.cos(action_angle), math.sin(action_angle)])
            current_direction = current / current_speed
            cos_angle = np.dot(action_vector, current_direction)
            r_current = cos_angle * (current_speed / self.auv_speed)
        else:
            r_current = 0
        
        # Компонент награды за плавность движения
        r_smoothness = 0
        if self.prev_action is not None:
            prev_angle = 2 * math.pi * self.prev_action / self.num_actions
            curr_angle = 2 * math.pi * action / self.num_actions
            angle_diff = abs(curr_angle - prev_angle)
            while angle_diff > math.pi:
                angle_diff = 2 * math.pi - angle_diff
            r_smoothness = math.cos(angle_diff)
        
        # Штраф за каждый шаг
        r_step = -2
        
        # Награда за достижение цели
        r_goal = 50 if np.linalg.norm(self.position - self.goal_pos) < 2.0 else 0
        
        # Композитная награда в соответствии с формулой
        total_reward = (
            5 * r_distance -      # Поощрение движения к цели
            8 * r_obstacles +     # Штраф за близость к препятствиям
            3 * r_current +       # Использование течений
            2 * r_smoothness +    # Плавность движения
            2 * r_step +          # Штраф за шаг
            r_goal                # Награда за достижение цели
        )
        
        return total_reward
    
    def render(self, figure=None, episode=None, show=True):
        """Визуализация среды"""
        if figure is None:
            plt.figure(figsize=(12, 8))
        else:
            plt.figure(figure.number)
            plt.clf()
        
        # Заголовок
        if episode is not None:
            plt.title(f'AUV Navigation - Episode {episode}')
        else:
            plt.title('AUV Navigation')
        
        # Отображение границ области
        plt.xlim(-5, self.area_width + 5)
        plt.ylim(-5, self.area_height + 5)
        plt.grid(True, linestyle='--', alpha=0.7)
        
        # Отображение препятствий
        for obstacle_pos in self.obstacles:
            circle = plt.Circle(
                (obstacle_pos[0], obstacle_pos[1]),
                self.obstacle_radius,
                color='red',
                alpha=0.5
            )
            plt.gca().add_patch(circle)
        
        # Отображение стартовой и целевой точек
        plt.scatter(self.start_pos[0], self.start_pos[1], color='green', s=100, label='Start')
        plt.scatter(self.goal_pos[0], self.goal_pos[1], color='blue', s=100, label='Goal')
        
        # Отображение траектории AUV
        trajectory_array = np.array(self.trajectory)
        plt.plot(trajectory_array[:, 0], trajectory_array[:, 1], 'k--', alpha=0.6)
        
        # Отображение текущей позиции AUV
        plt.scatter(self.position[0], self.position[1], color='black', s=80, label='AUV')
        
        # Отображение направления AUV
        heading_x = self.position[0] + 2 * math.cos(self.heading)
        heading_y = self.position[1] + 2 * math.sin(self.heading)
        plt.plot([self.position[0], heading_x], [self.position[1], heading_y], 'k-', linewidth=2)
        
        # Отображение показаний сонаров
        sonar_readings = self._get_sonar_readings()
        for i in range(self.num_sonars):
            angle = self.heading + (i - (self.num_sonars - 1) / 2) * math.radians(self.sonar_angle)
            dx = sonar_readings[i] * math.cos(angle)
            dy = sonar_readings[i] * math.sin(angle)
            plt.plot([self.position[0], self.position[0] + dx], 
                     [self.position[1], self.position[1] + dy], 
                     'g-', alpha=0.3)
        
        # Отображение поля течений (стрелки)
        x_grid = np.linspace(0, self.area_width, 10)
        y_grid = np.linspace(0, self.area_height, 10)
        X, Y = np.meshgrid(x_grid, y_grid)
        U = np.zeros_like(X)
        V = np.zeros_like(Y)
        
        for i in range(X.shape[0]):
            for j in range(X.shape[1]):
                current = self._get_current_at_position(np.array([X[i, j], Y[i, j]]))
                U[i, j] = current[0] * 5  # масштабирование для визуализации
                V[i, j] = current[1] * 5  # масштабирование для визуализации
        
        plt.quiver(X, Y, U, V, color='cyan', alpha=0.5, scale=10)
        
        # Добавление информации о текущем состоянии
        distance_to_goal = np.linalg.norm(self.position - self.goal_pos)
        info_text = f"Steps: {self.steps}\nDistance to goal: {distance_to_goal:.2f}"
        plt.figtext(0.02, 0.02, info_text, fontsize=10)
        
        plt.xlabel('X (морские мили)')
        plt.ylabel('Y (морские мили)')
        plt.legend()
        
        if show:
            plt.tight_layout()
            plt.draw()
            plt.pause(0.01)
        
        return plt.gcf()

import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import random
import math
import os
from collections import deque
import time
from tqdm import tqdm
import matplotlib.patches as patches

gpus = tf.config.experimental.list_physical_devices('GPU')

# tf.config.optimizer.set_jit(False)



# Включаем использование смешанной точности для ускорения
# mixed_precision = tf.keras.mixed_precision.experimental.Policy('mixed_float16')
# tf.keras.mixed_precision.experimental.set_global_policy(mixed_precision)
# print(f"Политика точности вычислений: {mixed_precision.name}")

# [Оставшийся код AUVEnvironment остается прежним]

# Класс NoisyDense - параметризованный шумовой слой для ND3QN
class NoisyDense(tf.keras.layers.Layer):
    def __init__(self, units, activation=None, sigma_init=0.017, name=None):
        super(NoisyDense, self).__init__(name=name)
        self.units = units
        self.activation = activation
        self.sigma_init = sigma_init
        
    def build(self, input_shape):
        # Определение размера входа
        self.input_dim = int(input_shape[-1])
        
        # Инициализация весов и смещений
        mu_range = 1 / np.sqrt(self.input_dim)
        
        # Средние веса
        self.mu_w = self.add_weight(
            name='mu_w',
            shape=[self.input_dim, self.units],
            initializer=tf.keras.initializers.RandomUniform(-mu_range, mu_range),
            trainable=True
        )
        self.mu_b = self.add_weight(
            name='mu_b',
            shape=[self.units],
            initializer=tf.keras.initializers.RandomUniform(-mu_range, mu_range),
            trainable=True
        )
        
        # Сигма для шума весов
        self.sigma_w = self.add_weight(
            name='sigma_w',
            shape=[self.input_dim, self.units],
            initializer=tf.constant_initializer(self.sigma_init / np.sqrt(self.input_dim)),
            trainable=True
        )
        self.sigma_b = self.add_weight(
            name='sigma_b',
            shape=[self.units],
            initializer=tf.constant_initializer(self.sigma_init / np.sqrt(self.input_dim)),
            trainable=True
        )
        
        # Необучаемые шумовые переменные
        self.epsilon_w = self.add_weight(
            name='epsilon_w',
            shape=[self.input_dim, self.units],
            initializer=tf.zeros_initializer(),
            trainable=False
        )
        self.epsilon_b = self.add_weight(
            name='epsilon_b',
            shape=[self.units],
            initializer=tf.zeros_initializer(),
            trainable=False
        )
        
        self.built = True
    
    def call(self, inputs, training=None):
        # В режиме обучения генерируем новый шум
        if training:
            self._sample_noise()
        
        # Применяем параметризованный шум
        weights = self.mu_w + self.sigma_w * self.epsilon_w
        bias = self.mu_b + self.sigma_b * self.epsilon_b
        
        # Линейное преобразование
        output = tf.matmul(inputs, weights) + bias
        
        # Применяем активацию, если указана
        if self.activation is not None:
            output = self.activation(output)
        
        return output
    
    def _sample_noise(self):
        """Метод для генерации нового шума"""
        # Функция для генерации факторизованного гауссовского шума
        def f(x):
            return tf.math.sign(x) * tf.math.sqrt(tf.math.abs(x))
        
        # Генерируем случайные переменные
        p = tf.random.normal([self.input_dim, 1])
        q = tf.random.normal([1, self.units])
        f_p = f(p)
        f_q = f(q)
        
        # Применяем факторизацию для весов
        epsilon_w = tf.matmul(f_p, f_q)
        
        # Обновляем шумовые переменные
        self.epsilon_w.assign(epsilon_w)
        self.epsilon_b.assign(f_q[0])


class ND3QN:
    def __init__(self, state_size, action_size, learning_rate=0.01, gamma=0.9, epsilon=0.8,
                 epsilon_decay=0.995, epsilon_min=0.01, update_target_freq=10,
                 batch_size=2048, memory_size=10000000, sigma=0.017, model_dir="models"):
        self.state_size = state_size
        self.action_size = action_size
        self.learning_rate = learning_rate
        self.gamma = gamma
        self.epsilon = epsilon
        self.epsilon_decay = epsilon_decay
        self.epsilon_min = epsilon_min
        self.update_target_freq = update_target_freq
        self.batch_size = batch_size
        self.memory = deque(maxlen=memory_size)
        self.sigma = sigma
        self.model_dir = model_dir
        
        # Счетчик для обновления целевой сети
        self.target_counter = 0
        self.total_updates = 0
        self.silent_updates = True  # Флаг для контроля вывода сообщений
        
        # Создание директории для сохранения моделей
        if not os.path.exists(model_dir):
            os.makedirs(model_dir)
        
        # Создание модели и целевой модели
        self.model = self._build_model()
        self.target_model = self._build_model()
        
        # Компилируем основную модель
        optimizer = tf.keras.optimizers.Adam(learning_rate=self.learning_rate)
        self.model.compile(optimizer=optimizer, loss='mse')
        
        # Логирование информации о модели
        self.model.summary()
        
        # Синхронизируем веса
        self.update_target_model(verbose=True)
        
    def _build_model(self):
        """Построение архитектуры модели ND3QN с оптимизациями"""
        # Входной слой
        inputs = tf.keras.layers.Input(shape=(self.state_size,))
        
        # Скрытые слои - упрощаем для уменьшения нагрузки
        x = tf.keras.layers.Dense(64, activation='relu', name='hidden1')(inputs)
        x = tf.keras.layers.Dense(64, activation='relu', name='hidden2')(x)
        
        # Noisy Layer
        noisy = NoisyDense(64, name='noisy')(x)
        
        # Разделение на две ветви: значение состояния (V) и преимущества действий (A)
        value_stream = NoisyDense(1, name='value_stream')(noisy)
        advantage_stream = NoisyDense(self.action_size, name='advantage_stream')(noisy)
        
        # Объединение ветвей Dueling DQN
        def combine_streams(inputs):
            value, advantage = inputs
            return value + (advantage - tf.reduce_mean(advantage, axis=1, keepdims=True))
        
        outputs = tf.keras.layers.Lambda(combine_streams, name='q_values')([value_stream, advantage_stream])
        
        # Построение модели
        model = tf.keras.models.Model(inputs=inputs, outputs=outputs)
        
        return model
    
    def update_target_model(self, verbose=False):
        """Обновление весов целевой модели"""
        self.target_model.set_weights(self.model.get_weights())
        self.total_updates += 1
        if verbose and not self.silent_updates:
            print(f"Целевая модель обновлена (#{self.total_updates})")
    
    def remember(self, state, action, reward, next_state, done):
        """Добавление перехода в память"""
        self.memory.append((state, action, reward, next_state, done))
    
    def act(self, state, training=True):
        """Выбор действия на основе текущего состояния"""
        if training and np.random.rand() <= self.epsilon:
            return random.randrange(self.action_size)
        
        state = np.expand_dims(state, axis=0)
        q_values = self.model.predict(state, verbose=0)[0]
        
        return np.argmax(q_values)
    
    def replay(self):
        """Обучение на батче опыта из памяти"""
        if len(self.memory) < self.batch_size:
            return 0
        
        # Выбор случайного батча
        minibatch = random.sample(self.memory, self.batch_size)
        
        # Подготовка данных
        states = np.zeros((self.batch_size, self.state_size))
        next_states = np.zeros((self.batch_size, self.state_size))
        actions, rewards, dones = [], [], []
        
        for i, (state, action, reward, next_state, done) in enumerate(minibatch):
            states[i] = state
            next_states[i] = next_state
            actions.append(action)
            rewards.append(reward)
            dones.append(done)
        
        # Предсказания для текущих и следующих состояний
        targets = self.model.predict(states, verbose=0)
        next_q_values = self.model.predict(next_states, verbose=0)
        target_q_values = self.target_model.predict(next_states, verbose=0)
        
        for i, (action, reward, done) in enumerate(zip(actions, rewards, dones)):
            if done:
                targets[i, action] = reward
            else:
                next_action = np.argmax(next_q_values[i])
                targets[i, action] = reward + self.gamma * target_q_values[i, next_action]
        
        # Обучение модели
        try:
            history = self.model.fit(
                states,
                targets,
                batch_size=self.batch_size,
                epochs=1,
                verbose=0
            )
            
            # Обновление целевой модели
            self.target_counter += 1
            if self.target_counter >= self.update_target_freq:
                self.update_target_model(verbose=False)
                self.target_counter = 0
            
            # Уменьшение эпсилон
            if self.epsilon > self.epsilon_min:
                self.epsilon *= self.epsilon_decay
            
            return history.history['loss'][0]
        except Exception as e:
            print(f"Ошибка при обучении: {e}")
            return 0
    
    def load(self, name=None):
        """Загрузка весов модели"""
        if name is None:
            name = os.path.join(self.model_dir, "nd3qn_auv_model.h5")
        try:
            self.model.load_weights(name)
            self.update_target_model(verbose=True)
            print(f"Модель загружена из {name}")
            return True
        except:
            print(f"Не удалось загрузить модель из {name}")
            return False
    
    def save(self, name=None):
        """Сохранение весов модели"""
        if name is None:
            name = os.path.join(self.model_dir, "nd3qn_auv_model.h5")
        self.model.save_weights(name)
        print(f"Модель сохранена в {name}")
# Функция визуализации результатов обучения
def plot_training_results(rewards, q_values, losses, success_rates, steps, save_path=None):
    """Визуализация результатов обучения"""
    plt.figure(figsize=(18, 12))
    
    # График наград
    plt.subplot(3, 2, 1)
    plt.plot(rewards)
    if len(rewards) > 10:
        plt.plot(np.convolve(rewards, np.ones(10)/10, mode='valid'), 'r-', alpha=0.7)
    plt.title('Награды за эпизоды')
    plt.xlabel('Эпизод')
    plt.ylabel('Награда')
    plt.grid(True)
    
    # График средних Q-значений
    plt.subplot(3, 2, 2)
    plt.plot(q_values)
    if len(q_values) > 10:
        plt.plot(np.convolve(q_values, np.ones(10)/10, mode='valid'), 'r-', alpha=0.7)
    plt.title('Средние Q-значения')
    plt.xlabel('Эпизод')
    plt.ylabel('Q-значение')
    plt.grid(True)
    
    # График функции потерь
    plt.subplot(3, 2, 3)
    plt.plot(losses)
    if len(losses) > 10:
        plt.plot(np.convolve(losses, np.ones(10)/10, mode='valid'), 'r-', alpha=0.7)
    plt.title('Функция потерь при обучении')
    plt.xlabel('Эпизод')
    plt.ylabel('Loss')
    plt.grid(True)
    
    # График успешности
    plt.subplot(3, 2, 4)
    plt.plot(success_rates)
    plt.title('Уровень успеха (скользящее окно 100 эпизодов)')
    plt.xlabel('Эпизод')
    plt.ylabel('Уровень успеха')
    plt.grid(True)
    
    # График числа шагов
    plt.subplot(3, 2, 5)
    plt.plot(steps)
    if len(steps) > 10:
        plt.plot(np.convolve(steps, np.ones(10)/10, mode='valid'), 'r-', alpha=0.7)
    plt.title('Число шагов за эпизод')
    plt.xlabel('Эпизод')
    plt.ylabel('Шаги')
    plt.grid(True)
    
    # Сравнение награды и успешности
    plt.subplot(3, 2, 6)
    plt.plot(rewards, label='Награда')
    
    # Нормализуем успешность для отображения на графике
    if len(rewards) > 0 and len(success_rates) > 0:
        max_reward = max(1, max(rewards))
        scaled_success = [s * max_reward for s in success_rates]
        plt.plot(scaled_success, 'g-', label='Успешность (масштаб)')
    
    plt.title('Сравнение награды и успешности')
    plt.xlabel('Эпизод')
    plt.ylabel('Значение')
    plt.legend()
    plt.grid(True)
    
    plt.tight_layout()
    
    if save_path:
        plt.savefig(save_path)
    
    plt.show()


# Функция обучения агента
def train_agent(episodes=1000, render_freq=10, save_freq=100, model_dir="models", model_name="nd3qn_auv_model"):
    """Функция для обучения агента"""
    # Создание полного пути для сохранения модели
    if not os.path.exists(model_dir):
        os.makedirs(model_dir)
    model_path = os.path.join(model_dir, f"{model_name}.h5")
    
    # Создаем среду
    env = AUVEnvironment()
    
    # Получаем размерности состояния и действий
    state_size = len(env.get_state())
    action_size = env.num_actions
    
    print(f"Размерность состояния: {state_size}")
    print(f"Размерность действий: {action_size}")
    
    # Создаем агента с оптимизированными параметрами
    agent = ND3QN(
        state_size=state_size,
        action_size=action_size,
        learning_rate=0.002,
        gamma=0.95,
        epsilon=0.8,
        epsilon_decay=0.99,
        epsilon_min=0.01,
        update_target_freq=10,
        batch_size=512,  # Уменьшаем размер батча для стабильности
        memory_size=100000,
        sigma=0.017,
        model_dir=model_dir
    )
    
    # Отключаем вывод обновлений целевой модели
    agent.silent_updates = True
    
    # Проверяем наличие сохраненной модели
    if os.path.exists(model_path):
        print(f"Загрузка сохраненной модели из {model_path}")
        agent.load(model_path)
    
    # Подготовка для визуализации
    fig = plt.figure(figsize=(12, 8))
    
    # Массивы для хранения статистики обучения
    episode_rewards = []
    avg_q_values = []
    loss_values = []
    success_rates = []
    avg_steps = []
    
    # Количество успешных эпизодов в последних 100
    success_window = deque(maxlen=100)
    
    # Директория для сохранения визуализаций
    vis_dir = os.path.join(model_dir, "visualizations")
    if not os.path.exists(vis_dir):
        os.makedirs(vis_dir)
    
    # Время начала обучения
    start_time = time.time()
    
    # Процесс обучения
    for episode in range(1, episodes + 1):
        print(f"\nНачало эпизода {episode}/{episodes}")
        sys.stdout.flush()  # Принудительный вывод
        
        # Сброс среды
        state = env.reset()
        
        total_reward = 0
        q_values_sum = 0
        loss_sum = 0
        loss_count = 0
        done = False
        step_count = 0
        
        # Один эпизод
        while not done:
            # Выбор действия
            action = agent.act(state)
            
            # Выполнение действия
            next_state, reward, done, info = env.step(action)
            
            # Сохранение опыта
            agent.remember(state, action, reward, next_state, done)
            
            # Обучение на батче опыта
            loss = agent.replay()
            
            # Обновление статистики
            state = next_state
            total_reward += reward
            step_count += 1
            
            # Сбор статистики для мониторинга (реже для оптимизации)
            if step_count % 20 == 0:  # Еще реже собираем статистику
                try:
                    q_values = agent.model.predict(np.expand_dims(state, axis=0), verbose=0)[0]
                    q_values_sum += np.mean(q_values)
                except Exception as e:
                    print(f"Ошибка при сборе статистики: {e}")
            
            if loss > 0:
                loss_sum += loss
                loss_count += 1
            
            # Периодический вывод прогресса внутри эпизода
            if step_count % 100 == 0:
                print(f"Эпизод {episode}, шаг {step_count}, reward: {total_reward:.2f}", end='\r')
                sys.stdout.flush()
        
        # Отслеживание успешных эпизодов
        success = info.get('goal_reached', False)
        success_window.append(1 if success else 0)
        success_rate = sum(success_window) / len(success_window)
        
        # Сохранение статистики эпизода
        episode_rewards.append(total_reward)
        avg_q_value = q_values_sum / max(1, step_count // 20)
        avg_q_values.append(avg_q_value)
        avg_loss = loss_sum / max(1, loss_count)
        loss_values.append(avg_loss)
        success_rates.append(success_rate)
        avg_steps.append(info['steps'])
        
        # Вывод информации о прогрессе обучения в конце эпизода
        elapsed_time = time.time() - start_time
        print(f"\nЭпизод {episode}/{episodes} завершен - Время: {elapsed_time:.1f}с")
        print(f"Награда: {total_reward:.2f}, Шаги: {info['steps']}, Успех: {success}")
        print(f"Эпсилон: {agent.epsilon:.4f}, Ср. Q-value: {avg_q_value:.4f}, Ср. Loss: {avg_loss:.6f}")
        print(f"Уровень успеха (посл. {len(success_window)}): {success_rate:.2%}, Дист. до цели: {info['distance']:.2f}")
        sys.stdout.flush()  # Принудительный вывод
        
        # Визуализация среды каждые render_freq эпизодов
        if episode % render_freq == 0:
            try:
                env.render(figure=fig, episode=episode)
                plt.savefig(os.path.join(vis_dir, f"episode_{episode}.png"))
                plt.pause(0.1)
            except Exception as e:
                print(f"Ошибка при визуализации: {e}")
        
        # Сохранение модели каждые save_freq эпизодов
        if episode % save_freq == 0:
            agent.save(model_path)
            try:
                plot_training_results(episode_rewards, avg_q_values, loss_values, success_rates, avg_steps, 
                                    save_path=os.path.join(vis_dir, "training_results.png"))
            except Exception as e:
                print(f"Ошибка при сохранении графиков: {e}")
    
    # Сохранение финальной модели
    agent.save(model_path)
    print(f"Финальная модель сохранена в {model_path}")
    
    # Визуализация результатов обучения
    try:
        plot_training_results(episode_rewards, avg_q_values, loss_values, success_rates, avg_steps,
                            save_path=os.path.join(vis_dir, "final_training_results.png"))
    except Exception as e:
        print(f"Ошибка при визуализации итогов: {e}")
    
    return agent, env

# Функция тестирования агента
def test_agent(agent, env, episodes=10, render=True, render_delay=0.5, vis_dir="visualizations"):
    """Функция для тестирования обученного агента"""
    if not os.path.exists(vis_dir):
        os.makedirs(vis_dir)
    
    success_count = 0
    total_rewards = 0
    total_steps = 0
    
    for episode in range(1, episodes + 1):
        state = env.reset()
        total_reward = 0
        steps = 0
        done = False
        
        # Подготовка для визуализации
        if render:
            fig = plt.figure(figsize=(12, 8))
        
        while not done:
            # Выбор действия без исследования (epsilon = 0)
            action = agent.act(state, training=False)
            
            # Выполнение действия
            next_state, reward, done, info = env.step(action)
            
            # Обновление статистики
            state = next_state
            total_reward += reward
            steps += 1
            
            # Рендеринг, если требуется
            if render and steps % 5 == 0:  # рендеринг каждые 5 шагов для ускорения
                env.render(figure=fig, episode=episode)
                plt.savefig(os.path.join(vis_dir, f"test_episode_{episode}_step_{steps}.png"))
                plt.pause(render_delay)
        
        # Финальный рендеринг эпизода
        if render:
            env.render(figure=fig, episode=episode)
            plt.savefig(os.path.join(vis_dir, f"test_episode_{episode}_final.png"))
            plt.pause(render_delay)
        
        # Обновление статистики
        success = info.get('goal_reached', False)
        if success:
            success_count += 1
        
        total_rewards += total_reward
        total_steps += steps
        
        print(f"Тест, эпизод {episode}: Награда = {total_reward:.2f}, Шаги = {steps}, Успех = {success}")
    
    # Вывод итоговой статистики
    avg_reward = total_rewards / episodes
    avg_steps = total_steps / episodes
    success_rate = success_count / episodes
    
    print(f"\nРезультаты тестирования:")
    print(f"Средняя награда: {avg_reward:.2f}")
    print(f"Среднее число шагов: {avg_steps:.2f}")
    print(f"Уровень успеха: {success_rate:.2%}")
    
    # Сохранение статистики в файл
    with open(os.path.join(vis_dir, "test_results.txt"), 'w') as f:
        f.write(f"Результаты тестирования ({episodes} эпизодов):\n")
        f.write(f"Средняя награда: {avg_reward:.2f}\n")
        f.write(f"Среднее число шагов: {avg_steps:.2f}\n")
        f.write(f"Уровень успеха: {success_rate:.2%}\n")
    
    return avg_reward, avg_steps, success_rate


# Основная функция
def main():
    # Проверка наличия CUDA
    if tf.config.list_physical_devices('GPU'):
        print("CUDA доступен. Обучение будет выполняться на GPU.")
    else:
        print("CUDA недоступен. Обучение будет выполняться на CPU.")
    
    # Создание директорий для результатов
    model_dir = "auv_models"
    vis_dir = os.path.join(model_dir, "visualizations")
    
    if not os.path.exists(model_dir):
        os.makedirs(model_dir)
    if not os.path.exists(vis_dir):
        os.makedirs(vis_dir)
    
    # Параметры обучения
    episodes = 1000
    render_freq = 10
    save_freq = 100
    model_name = "nd3qn_auv_model"
    
    # Обучение агента
    agent, env = train_agent(episodes, render_freq, save_freq, model_dir, model_name)
    
    # Тестирование агента
    test_agent(agent, env, episodes=10, render=True, vis_dir=vis_dir)


if __name__ == "__main__":
    main()
```
